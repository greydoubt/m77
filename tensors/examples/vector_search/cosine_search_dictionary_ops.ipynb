Pickling is the process of converting a Python object into a byte stream, while unpickling is the reverse operation, converting a byte stream back into a Python object.

Here's a simple demonstration of how to use the pickle module to pickle and unpickle Python objects:

    We'll create a dictionary.
    We'll pickle the dictionary and save it to a file.
    We'll read the file and unpickle the content to get the original dictionary back.

Here's the code:

```python
import pickle

# 1. Create a dictionary
data = {
    "name": "John",
    "age": 25,
    "is_student": False
}

# 2. Pickle and save the dictionary to a file
with open("data.pkl", "wb") as file:
    pickle.dump(data, file)

# 3. Read the file and unpickle the content
with open("data.pkl", "rb") as file:
    loaded_data = pickle.load(file)

print(loaded_data)
```

Packaging functions in a pickle and then loading them is a more advanced and potentially risky use of the pickle module. This is because pickling functions or code in general can expose security vulnerabilities if not handled with care.

However, for educational purposes, here's how you can pickle a function and then load it:

    Define a function.
    Pickle and save the function to a file.
    Load the function from the file and execute it.

Let's define a simple function that adds two numbers, pickle it, and then load and execute it:

```python
import pickle

def add_numbers(a, b):
    return a + b

# Pickle and save the function to a file
with open("function.pkl", "wb") as file:
    pickle.dump(add_numbers, file)

# Load the function from the file
with open("function.pkl", "rb") as file:
    loaded_function = pickle.load(file)

# Execute the loaded function
result = loaded_function(10, 5)
print(result)
```

    Generate Word Embeddings: We'll create a mock dictionary of word embeddings for simplicity. In a real-world scenario, you might use models like Word2Vec, FastText, or GloVe to generate these embeddings.
    Pickle the Dictionary: We'll pickle and save the dictionary with word embeddings.
    Load the Pickled Dictionary: We'll load the pickled dictionary.
    Compute Cosine Similarity: Given a query word embedding, we'll search for the most similar word in our dictionary based on cosine similarity.

```python
import numpy as np

# Mock dictionary of word embeddings
word_embeddings = {
    "apple": np.array([0.1, 0.2, 0.3]),
    "banana": np.array([0.3, 0.2, 0.1]),
    "cherry": np.array([0.5, 0.5, 0.5]),
    "date": np.array([0.7, 0.8, 0.2])
}

# Pickle and save the dictionary
embeddings_path = "word_embeddings.pkl"
with open(embeddings_path, "wb") as file:
    pickle.dump(word_embeddings, file)

embeddings_path
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Load the pickled dictionary of word embeddings
with open(embeddings_path, "rb") as file:
    loaded_embeddings = pickle.load(file)

def find_most_similar(query_word, embeddings):
    """
    Given a query word and dictionary of embeddings, 
    returns the most similar word based on cosine similarity.
    """
    if query_word not in embeddings:
        return "Word not in dictionary."
    
    query_embedding = embeddings[query_word].reshape(1, -1)
    max_similarity = -float("inf")
    most_similar_word = None
    
    for word, embedding in embeddings.items():
        if word == query_word:
            continue
        similarity = cosine_similarity(query_embedding, embedding.reshape(1, -1))
        if similarity > max_similarity:
            max_similarity = similarity
            most_similar_word = word
            
    return most_similar_word

# Find the word most similar to "apple"
similar_word_to_apple = find_most_similar("apple", loaded_embeddings)
similar_word_to_apple
```

    Generate Sentence Embeddings: Create a mock dictionary of sentence embeddings for a set of example sentences.
    Pickle the Dictionary: We'll pickle and save the dictionary with sentence embeddings.
    Load the Pickled Dictionary: We'll load the pickled dictionary.
    Compute Cosine Similarity: Given a query sentence embedding, we'll search for the most similar sentence in our dictionary based on cosine similarity.

```python
# Mock dictionary of sentence embeddings
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Machine learning is transforming industries.",
    "The sun rises in the east and sets in the west.",
    "The mountains are beautiful during winter.",
    "I enjoy reading books on rainy days."
]

# Generate random embeddings for each sentence in a 300-dimensional space
sentence_embeddings = {sentence: np.random.rand(300) for sentence in sentences}

# Pickle and save the dictionary
sentence_embeddings_path = "sentence_embeddings.pkl"
with open(sentence_embeddings_path, "wb") as file:
    pickle.dump(sentence_embeddings, file)

sentence_embeddings_path
```

```python
# Load the pickled dictionary of sentence embeddings
with open(sentence_embeddings_path, "rb") as file:
    loaded_sentence_embeddings = pickle.load(file)

def find_most_similar_sentence(query_sentence, embeddings):
    """
    Given a query sentence and dictionary of embeddings, 
    returns the most similar sentence based on cosine similarity.
    If the sentence isn't in the dictionary, a random embedding is generated for the query.
    """
    if query_sentence not in embeddings:
        query_embedding = np.random.rand(300).reshape(1, -1)
    else:
        query_embedding = embeddings[query_sentence].reshape(1, -1)
    
    max_similarity = -float("inf")
    most_similar_sentence = None
    
    for sentence, embedding in embeddings.items():
        if sentence == query_sentence:
            continue
        similarity = cosine_similarity(query_embedding, embedding.reshape(1, -1))
        if similarity > max_similarity:
            max_similarity = similarity
            most_similar_sentence = sentence
            
    return most_similar_sentence

# Find the sentence most similar to the query
query_sentence = "I love reading novels when it's raining."
similar_sentence = find_most_similar_sentence(query_sentence, loaded_sentence_embeddings)
similar_sentence
```

Note: this uses random values for vectors so it is not accurate!
